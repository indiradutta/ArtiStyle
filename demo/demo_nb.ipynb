## **Initializing the NST Model**

```python
from nst import NST
nst_res = NST(gpu=True)
```

## **Setting image paths and running inferences**

```python
content_img, style_img = ('content.jpg', 'style.jpg')
result = nst_res.run_nst(content_img, style_img)
```

## **Visualizing**

```python
import matplotlib.pyplot as plt

plt.imshow(result)
plt.show()
```
